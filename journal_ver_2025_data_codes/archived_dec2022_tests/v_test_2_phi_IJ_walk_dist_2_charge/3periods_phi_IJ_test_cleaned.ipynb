```python

def _get_data_from_solved_mpdp(
    solver_results,
    solved_model: pyo.ConcreteModel,    
    ) -> dict:
    """_summary_

    Args:
        solver_results (_type_): _description_
        solved_model (pyo.ConcreteModel): _description_

    Returns:
        dict: _description_
    """
    # Check solver_results, if model is unsolved, return Error.
    assert solver_results.solver.status == 'ok', '''Cannot save unsolved model results.'''
    
    # Initialize useful information to save
    useful_description_info_list = ['Lower bound', 'Upper bound', 'Number of constraints', 'Number of variables']
    useful_solver_info_list = ['User time', 'Termination message', 'Statistics', 'Time' ]
    # Get data from solved_model
    solved_model_data = _get_data_from_frame_model(frame_model = solved_model)

    # Get objective value from solved_model
    obj_components = [
        solved_model.obj_total_cost, 
        solved_model.cs_placement_cost, 
        solved_model.grid_expansion_cost, 
        solved_model.backstop_tech_cost_sp, 
        solved_model.gross_revenue,
        solved_model.cs_placement_cost + solved_model.grid_expansion_cost - solved_model.gross_revenue
    ]
    obj_components_name = [
        'obj_total_cost',
        'cs_placement_cost', 
        'grid_expansion_cost', 
        'backstop_tech_cost_sp', 
        'gross_revenue',
        'first_stage_cost',
    ]      
    obj_val_dict = {obj_components_name[idx]:pyo.value(obj_Expr) for idx, obj_Expr in enumerate(obj_components)}       
    
    # print("Storing test results ... ")        
    test_result_dict = {
        **obj_val_dict,
        # Store selected information from solver_results
        'prob_description': {
                key: solver_results['Problem'][0][key] for key in useful_description_info_list 
            },
        'solver_info' : {
                key: solver_results['Solver'][0][key] for key in useful_solver_info_list
            },
        'solved_model_data': solved_model_data #mpsp_decisions 
    }
    # print("Done!")
    return test_result_dict
```

```python
def CS_service_time_Cap_rule(model, i, t):             
    service_time_required = pyo.value(sum(
        model.z[i,j,t] * model.tau[i,j] * model.calA[j,t] * (model.calT[j,t]+2*model.d[i,j]) \
            for j in model.J
    ))
    return service_time_required <= pyo.value(model.DeltaT[t]*(model.y[i]+model.n[i]))
```
